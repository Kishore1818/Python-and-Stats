######  <span style="font-family: Arial; font-weight:bold;font-size:1.0em;color:#c3358a">Activation function: Binary Step
<font color=darkblue>   
    
$$
f(x) = \left\{
        \begin{array}{lll}
            0 & for & x \leq 0  \\
            1 & for & x > 0
        \end{array}
    \right.
$$
  
<br>
    
$$
f'(x) = \left\{
        \begin{array}{lll}
            0 & for & x \neq 0  \\
            ? & for & x = 0
        \end{array}
    \right.
$$

```python
x = np.arange(-6, 6, 0.01)
binary_step = np.vectorize(lambda x: 1 if x > 0 else 0, otypes=[np.float])
z = binary_step(x)

plt.plot(x, z)
plt.title("Binary Step",fontsize=15)
plt.grid(alpha=0.4, ls='-.')
plt.xlabel("$z$",fontsize=15)
plt.ylabel("$\sigma(z)$",fontsize=15)
```

######  <span style="font-family: Arial; font-weight:bold;font-size:1.0em;color:#9135c3">Activation function: Piecewise Linear
<font color=darkblue>   
    
$$
f(x) = \left\{
        \begin{array}{lll}
            0 & for & x < x_{min} \\
            mx+b & for & x_{min} \leq x \leq x_{max}  \\
            1 & for & x > x_{max}
        \end{array}
    \right.
$$
<br>
    
$$
f'(x) = \left\{
        \begin{array}{lll}
            0 & for & x < x_{min} \\
            m & for & x_{min} \leq x \leq x_{max}  \\
            0 & for & x > x_{max}
        \end{array}
    \right.
$$

```python
x = np.arange(-6, 6, 0.01)
piecewise_linear = np.vectorize(lambda x: 1 if x > 3 else 0 if x < -3 else 1/6*x+1/2, otypes=[np.float])
z = piecewise_linear(x)

plt.plot(x, z)
plt.title("Piecewise Linear",fontsize=15)
plt.axhline(lw=1, c='black')
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.4, ls='-.')
plt.xlabel("$z$",fontsize=15)
plt.ylabel("$\sigma(z)$",fontsize=15)
```
