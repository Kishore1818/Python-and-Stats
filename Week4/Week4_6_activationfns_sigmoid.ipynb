######  <span style="font-family: Arial; font-weight:bold;font-size:1.0em;color:#b735c3">Activation function: Sigmoid
<font color=darkblue>   

- It is used as the activation function in all layers of our network and is defined as:

$$\sigma(z) = \frac{1}{1 + e^{-z}}$$
    
<br>
$$
f'(z)=f(z)(1-f(z))
$$

- where $z$ is an `ndarray` (vector) and it will apply the formula element-wise and return another `ndarray` of the same shape as $z$.

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
# z = 1/(1 + np.exp(-x))
z = sigmoid(x)
# print(z)
plt.plot(x, z)
plt.title("Sigmoid",fontsize=15)
plt.xlabel("$z$",fontsize=15)
plt.ylabel("$\sigma(z)$",fontsize=15)
plt.grid(alpha=0.4, ls='-.')
plt.text(-8.5,0.85,r'$ \sigma(z) = \frac{1}{1+e^{-z}}$',fontsize=15)
```

```python
def sigmoid_prime(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

```python
x = np.linspace(-10, 10, 100)
z = sigmoid_prime(x)
# print(z)
plt.plot(x, z)
plt.xlabel("$z$",fontsize=15)
plt.ylabel("$\sigma(z)$",fontsize=15)
```
