######  <span style="font-family: Arial; font-weight:bold;font-size:1.0em;color:#4335c3">Activation function: TanH
<font color=darkblue>   

```python
%matplotlib inline
x = np.arange(-6, 6, 0.05)
y = np.tanh(x)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title("TanH",fontsize=15)
ax.set_xlabel("$z$",fontsize=15)
ax.set_ylabel("$\sigma(z)$",fontsize=15)
plt.grid(alpha=0.4, ls='-.')
ax.text(-6.5,0.8,r'$ \sigma(z) = \!\mathrm{tanh}(z)$',fontsize=15)
```

######  <span style="font-family: Arial; font-weight:bold;font-size:1.0em;color:#c3bc35">Activation function: Arctangent, ArcTan
<font color=darkblue>  
    
$$
f(x)=tan^{-1}(x)
$$
    
<br>
$$
f'(x)={\frac {1}{1+x^2}}
$$

```python
def arctan(x):
    return np.arctan(x)

%matplotlib inline
x = np.arange(-6, 6, 0.05)
y = arctan(x)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title("Arctangent",fontsize=15)
ax.set_xlabel("$z$",fontsize=15)
ax.set_ylabel("$\sigma(z)$",fontsize=15)
plt.grid(alpha=0.4, ls='-.')
```
