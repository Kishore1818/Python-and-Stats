######  <span style="font-family: Arial; font-weight:bold;font-size:1.0em;color:#4335c3">Activation function: Leaky ReLU
<font color=darkblue>   

```python
# The following function is taken from Weidman (2019):
def leaky_relu(x: ndarray) -> ndarray:
    """
    Apply Leaky ReLU to each element in ndarray.
    """
    return np.maximum(0.2 * x, x)
```

```python
%matplotlib inline
x = np.arange(-2, 2, 0.05)
y = leaky_relu(x)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title("Leaky ReLU",fontsize=15)
ax.set_xlabel("$z$",fontsize=15)
ax.set_ylabel("$\sigma(z)$",fontsize=15)
```

######  <span style="font-family: Arial; font-weight:bold;font-size:1.0em;color:#358ac3">Activation function: ReLU
<font color=darkblue>   

```python
%matplotlib inline
x = np.arange(-2, 2, 0.05)
y = x.clip(min=0)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title("ReLU",fontsize=15)
ax.set_xlabel("$z$",fontsize=15)
ax.set_ylabel("$\sigma(z)$",fontsize=15)
ax.text(-2,1.7,r'$\sigma (z) = \mathrm{max}(z,0)$',fontsize=15)
```
