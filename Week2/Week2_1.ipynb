- _Import Libraries_

```python
import pandas as pd
import seaborn as sns   # Why sns?  It's a reference to The West Wing
import matplotlib.pyplot as plt  # seaborn is based on matplotlib
sns.set(color_codes=True) # adds a nice background to the graphs
%matplotlib inline 
```

- _Data reading, display some rows and all columns_

```python
df = pd.read_csv('/home/jayanthikishore/Downloads/ML_classwork/Week2/Automobile.csv')  #linux or mac (just change direction of '/' for win)
df.head()
```

- _Display end rows and all columns_

```python
df.tail()
# df.tail(3) #just it will display 3 rows and columns
```

- __Univariate distributions: By default, this will draw a histogram and fit a kernel density estimate (KDE).__

```python
sns.set({'figure.figsize': (8,6)})
sns.histplot(df['highway_mpg'], kde=True);
#if you have any issues while plotting just install: "pip install -U seaborn"
```

- _without Kernel Density Estimate (KDE)_

```python
sns.set({'figure.figsize': (8,6)})
sns.histplot(df['city_mpg'], kde=False);
```

- __Bivariate distributions: Finding the relationship between two variables__

```python
sns.set({'figure.figsize': (8,8)})
sns.jointplot(x='engine_size', y='horsepower', data=df);
```

- __Hexabin representations of two variables__

```python
sns.jointplot(x='engine_size', y='horsepower', data=df, kind="hex");
```

- __Kernal Density Estimation__

```python
sns.jointplot(x='engine_size', y='horsepower', data=df, kind="kde");
```

- __Pairwise relationship between the variables__

```python
sns.pairplot(df[['normalized_losses', 'engine_size', 'horsepower']]);
```

- __Plotting with categorical data__
- _strip plot shows all data points in one place and use "jitter" and it will adjust the without overlapping_

```python
sns.stripplot(x='fuel_type', y='horsepower', data=df, jitter=True);
```

- __same as above but different approach as "Swarmplot" function__

```python
sns.swarmplot(x='fuel_type', y='horsepower', data=df);
```

- __Boxplot with quartiles__

```python
sns.boxplot(x='number_of_doors', y='horsepower', data=df);
```

```python
sns.boxplot(x='number_of_doors', y='horsepower', data=df, hue=df['fuel_type']);
```

- __Bar Plots__
- _We can plot the mean of a a dataset, separated in categories using the barplot() function. When there are multiple observations in each category, it uses bootstrapping to compute a confidence interval around the estimate and plots that using error bars:_

```python
sns.barplot(x='body_style', y='horsepower', data=df, hue=df['fuel_type']);
```

- __Bar Plot__
- _The number of observations in each category rather than computing the mean_

```python
sns.countplot(x='body_style', data=df, hue=df['fuel_type']);
```

- __Line plot with symbols__

```python
sns.pointplot(x='body_style', y='horsepower', data=df, hue=df['number_of_doors']);
```

- __Drawing multi-panel categorical plots__

```python
sns.catplot(x="fuel_type",
               y = "horsepower",
               hue="number_of_doors", 
               col="drive_wheels", 
               data=df, 
               kind="box");
```

- __Linear Regression Model fit with confidence levels__
- _finding the relation between the two variables_

```python
sns.lmplot(y="horsepower", x="engine_size", data=df);
```

```python
sns.lmplot(y="horsepower", x="engine_size",hue="fuel_type", data=df);
```
